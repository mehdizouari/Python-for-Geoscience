# KNN Model Summary

````python
# Load data
well = pd.read_csv(filename)
well = well.drop(['DEPTH'],1)
Non_NAN = well['RHOB'].notnull().values
well = well[Non_NAN]

# Divide data in predictors and labels
x = well[['RHOB', 'DT', 'GR']]
y = well['FACIES'].values

# Scale predictors (X)
x = preprocessing.scale(x)

# Split data in Test and Train
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 2)

######
k = 10 # Number of neighbors!!!!!!!!!!!!!!

# Define Model
KNN_classifier = neighbors.KNeighborsClassifier(n_neighbors = k)

# Train Model
KNN_classifier.fit(x_train, y_train)

# Apply model to Test data
y_pred_KNN = KNN_classifier.predict(x_test)

# Create classification report
print('Model KNN = ', k)
print(classification_report(y_test, y_pred_KNN))

# Create a confusion matrix and plot it
KNN_cm = confusion_matrix(y_test, y_pred_KNN)
plt.figure(figsize=(14,8))
sns.heatmap(KNN_cm, cmap = 'GnBu', annot = True, square = True)
plt.title('Model KNN = %i' %k, fontsize = 25)
plt.xlabel('Predicted')
plt.ylabel('Truth')
````

### PCA

````python
# Step 0: Import library
from sklearn.decomposition import PCA
#############
# Step 1: Load data
well = pd.read_csv(filename)
Non_NAN = well['RHOB'].notnull().values
well = well[Non_NAN]

#################################################################

# Divide data in predictors and labels
x = well[['RHOB', 'DT', 'GR']] # Here is where your INDEPENDENT select the variable
y = well['FACIES'].values # DEPENDENT variable

# Scale predictors (X)
x = preprocessing.scale(x)
#############
# Step 2: Define PCA model
pca = PCA(n_components=2, svd_solver='full')
````

````python
# Step 3: Apply the model to our data
pca.fit(x)
````

````python
# Step 4: Transform the data into the PCA domain
pca_results = pca.transform(x)
````

````python
# Step 5: Check results
print('ORIGINAL data shape:', x.shape)
print('PCA data shape:', pca_results.shape)
````
